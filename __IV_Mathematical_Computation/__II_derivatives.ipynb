## Išvestinės apibrėžimas

Išvestinė tai matematinė funkcija, rodanti tam tikros funkcijos pokyčio tempą tam tikrame taške. Tai viena iš dviejų pagrindinių integralinio ir diferencialinio skaičiavimų savokų. Vaizduojant funkciją kaip dvimatį grafiką, išvestinė tam tikrame taške gali būti vaizduojama kaip liestinės tame taške krypties koeficientas. Išvestines turi ne visos funkcijos.
Pažiūrėkime kas tai ir kaip rasti funkcijos išvestinę iš grafiko.

Turime tiese, kurios funkcija yra <code>**mx + b**</code>. m yra tiesės krypties koficientas dar vadinamas tiesės nuolydis yra lygus x ir y pokyčio santykiui

\begin{multline*}
\dfrac{\Delta y}{\Delta x}
\end{multline*}

Tarkime vieno taško kordinatės yra <code>**x0 y0**</code> ir kito taško kordinatės yra  <code>**x1 y1**</code>  tuomet Δ (pokyčiai) apskaičiuojami taip:

\begin{multline*}
\dfrac{y_{1} - y_{0}}{x_{1} - x_{0}} = \dfrac{6}{6} = 1
\end{multline*}

Turime kreivė f(x) = x². Tuomet mes galėtume lygiai tai pat kaip ir pirmų atveju surasti tos kreivės kirstinės (angl. secant) kuri irgi yra tiesė, krypties koeficientą. Brėžiame kreivės kirstinė per du kreivės taškus (A, B).

```python
%run ../scripts/x^2_derivative.py
```

Taško A kordinatės <code>**(x0, f(x0))**</code>, taško B kordinates <code>**(x1, f(x1))**</code>. Dabar galėsime rasti kirstinės einančios per du funkcijos taškus krypties koeficientą.  TODO: funkcijos ir argumento pokyčio nustatymas

\begin{multline*}
m = \dfrac{f(y_{1}) - f(y_{0})}{x_{1} - x_{0}} = \dfrac{\Delta f(x)}{\Delta x}
\end{multline*}

Taip randamas funkcijos kirstinės pokytis

Dabar kai mes tą kirstine suksime nuo pradinės padėties nejudindami taško A, o taška B suksime tai atsiras tokia padėtis kad ta <code>**kirstinė virs liestinė**</code>, šį kreivė jau nebe kirs o lies funkcijos grafiką. 

Kaip randama funkcijos grafiko kuri yra liestinė krypties koficientas. Turint tą pačia kreivę, pokytis kai kirstinė sukosi, pokytis Δ𝑥 artėjo į nulį. Principe liestinėje x0 beveik sutaps arba visiškai sutaps su x1. Liestinė irgi yra tiesė ir jos krypties koeficientas bus

\begin{multline*}
m = \dfrac{\Delta f(x)}{\Delta x} = \dfrac{0}{0}
\end{multline*}

Bet 0/0 operacija yra negalima todėl mes negalime surasti liestinės krypties koeficiento. Tai ką daryti ?
Priartinkime ar padidinkime ta zoną kurioje liestinė liečia kreivę ir tada liestinė vėl bus kirstinė nes šį kirstinė kertą funkciją be galo artimuose taškuose A ir B. Ir tik toje zonoje matysime kad kirstinė taps liestinė kai taško A nejudiname, o tašką B "leidžiame žemyn" tokiu budu šį kirstinė taip virs liestinė. Tada taškai A ir B visai sutaps. <code>**Kirstinės ribinė padėtis bus liestinė  kai Δ𝑥 → 0.**</code> 

Funkcijos f išvestinė taške x0 vadinamas skaičius prie kurio artėja santykis

\begin{multline*}

\end{multline*}

### Lagranžo and Leibniz notacija

\begin{multline*}
f′(x) = \dfrac{df}{dx} = \lim_{\Delta x \rightarrow 0} = \dfrac{f(x + \Delta x) - f(x)}{\Delta x}
\end{multline*}

<code>**Grandinės taisyklė**</code>

\begin{multline*}
f(x) = f(g(x)) = z
\end{multline*}

<code>**Lagranžo notacija:**</code>

\begin{multline*}
f(x)′ = f(g(x))′g(x)′ = z′
\end{multline*}

<code>**Leibniz notacija:**</code>

\begin{multline*}
\dfrac{d}{dx}[f(g(x))] = \dfrac{df}{dg} \cdot \dfrac{dg}{dx}
\end{multline*}

Pytorch kurs automatiškai tokias skaičaivimų sekas ir jeigu mes nustatysime šiuos skaičiavimus išsaugos. Pytorch skaičiuoja išvestines diferencijuojamoms funkcijoms automatiškai `autograd` pakuotės pagalba. Bet tai nereiškai kad mums nereikia žinoti kaip skaičiuojamos išvestinės.

\begin{multline*}
f(x) = \log(\sqrt{x})
\end{multline*}

\begin{multline*}
\dfrac{df}{dx} = \dfrac{d}{dg}log(g) \cdot \dfrac{d}{dx}\sqrt{x}
\end{multline*}

\begin{multline*}
\dfrac{d}{dg}log(g) = \dfrac{1}{g} = \dfrac{1}{\sqrt{x}} 
\end{multline*}

\begin{multline*}
\dfrac{d}{dx}x^{1/2} = \dfrac{1}{2}x^{-1/2} = \dfrac{1}{2\sqrt{x}} 
\end{multline*}

\begin{multline*}
\dfrac{df}{dx} = \dfrac{1}{\sqrt{x}} \cdot \dfrac{1}{2\sqrt{x}} = \dfrac{1}{2x}
\end{multline*}

Tarkime, kad jūs bandote parduoti konkretų bųta savo kaimynystėje ir šio buto parametrai gali pasakyti, už kiek galėtumėte parduoti savo turtą. Na, vienas dalykas, kurį galime padaryti, yra pritaikyti modelį. Nustatysime tiesią šių duomenų liniją ir pagal tai nuspėsime, kokią kainą turėtume nustatyti savo turtui.

```python
%run ../scripts/_linear_regression_sample_c.py
```

Hipotezės arba modelio išraiška, dažnai aprašoma skirtingai.Mes dažniausiai naudosime tai:

$$ y = mx +b $$

$$h_{\theta}(x) = \theta_{0} + \theta_{1}x$$
<br>
$$h_{\theta}(x)\rightarrow h(x)$$

<br>

Akivaizdu kad y yra tam tikra f(x) tiesė. <code>**Kodėl naudojame tiesinė funkcija**</code>, bent jau kol kas dėl to kad tai yra paprasta. Dažniausiai taikysime ir sudėtingesnes netiesines funkcijas. Tačiau kadangi šis linijinis atvejis yra <code>**pamatinis mūsų blokas**</code>, kuriuo remsimės, kad galiausiai sukurtume sudėtingesnius modelius. Priminimas, kad šis konkretus modelis vadinamas <code>**tiesine regresija su 1 kintamuoju, kai kintamasis yra x**</code> Numatomos visos kainos kaip vieno kintamojo x funkcija. Ir dažniausiai naudojamas šio modelio pavadinimas yra <code>**vienamatė (protingai skambantis vieno kintamojo pavadinimas) tiesinė regresija**</code>.

Turime uždavinį taikant šią tiesinę regresiją su konkrečių duomenų rinkiniu. Mes norime surasti koeficientų vertę, kad mūsų tiesinis modelis kažkaip gerai prisitaikytu prie mokymo duomenų.
Tačiau kaip sugalvoti tų koeficientų reikšmes, kad būtų geras atitikimas? Pagrindinė mintis yra ta, kad turime pasirinkti parametrą, kuris <code>**h(x) reiškia, kad reikšmė, kurią prognozuojame įvesdami x, yra bent artima mūsų mokymo rinkinio pavyzdžių reikšmėms y**</code>. Pateikiame keletą pavyzdžių, kai žinome (x – kaip buto dydį) ir žinome (y tikrąją jo kainą). Taigi pabandykime parinkti koeficientų parametrų reikšmes taip, kad bent treniruočių rinkinyje numatytų gerai y reiškmes. Mums reikės išspręsime vadinamąją <code>**minimumo problemą**</code> 

$$\min\theta_{0} \theta_{1}$$

Mes norime, kad skirtumas tarp modelio h(x) ir y būtų mažas. Ir vienas dalykas, kurį galime padaryti, tai pabandyti sumažinti <code>**kvadratinį skirtumą tarp hipotezės išvesties ir faktinės kainos**</code>.

$$ \hat{y} = h_{\theta}(x) = \theta_{0} + \theta_{1}x$$

$$(\hat{y} -y)^2$$

Šį klaidos atstumą galite įsivaizduoti kaip vertikalų atstumą tarp hipotezės ir y vertės (nuosavybės kainos), dar vadinamos <code>**likučiais (ang. residuals)**</code>

[grafikas](https://kwichmann.github.io/ml_sandbox/linear_regression_diagnostics/)

Mes turime susumuoti šias klaidas <code>**(į kvadratinių skirtumų sumą)**</code> tarp prognozės ir įvesties atėmus faktinę konkretaus turto kainą. Kad matematika vėliau būtų šiek tiek lengvesnė, sumažinti savo vidurkį su pastovia <code>**1/2**</code> reiškme priekyje, kad būtų patogiau apskaičiuoti kvadratinės funkcijos išvestinius terminus, juos tiesiog vėliau 1/2 reikšmė panaikins.

$$ \min\theta_{0} \theta_{1} \   \ \dfrac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)} ) -y^{(i)})^2$$


Tai yra mūsų tikslo funkcija tiesinei regresijai. Paprastai mes apibrėžiame <code>**kaštų funkciją**</code> arba <code>**vidutinės kvadratinės klaidos funkciją (angl. mean squared error function - MSE)**</code>. Pasirodo, imame šiuos klaidų kvadratus, nes tai yra pagrįstas pasirinkimas ir puikiai tinka daugeliui regresijos problemų, nors yra ir kitų kaštų funkcijų, kurios taip pat gerai veiks, tačiau ši yra dažniausiai naudojama.

$$ J(\theta_{0}, \theta_{1}) = \dfrac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)} ) -y^{(i)})^2$$

Kad ši lygtis būtų aiški, <code>**h apatinis indeksas theta**</code> yra:

$$ h_{\theta}(x) = \theta_{0} + \theta_{1}x$$

Mūsų optimizavimo tikslai yra sumažinti:

$$\min  J(\theta_{0}, \theta_{1})$$

Norėdami suprasti, kas yra kaštų funkcija ir kodėl norime ją naudoti, dirbkime su supaprastinta hipoteze, kuri turi tik <code>**vieną parametrą 𝜃1 , 𝜃0 bus lygus nuliui**</code>.

$$h_{\theta}(x) = \theta_{1}x, \theta_{0} = 0 $$


$$J(\theta_{1}) = \dfrac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)} ) -y^{(i)})^2 , h_{\theta}(x^{(i)}) = \theta_{1}x^{(i)}$$


Mūsų optimizavimo tikslai yra sumažinti
$$\min J(\theta_{1})$$

Grafike tai reiškia, kad jei 𝜃 0 = 0, tai atitinka tik hipotezės funkcijos, kuri praeina per <code>**(0,0)**</code> tašką. Taip supaprastintas kaštų funkcijos hipotezės apibrėžimas leis mums  ją geriau suprasti.

<code>**Pagrindinės funkcijos, apie kurias reikia galvoti**</code>:

\begin{multline*}
h_{\theta}(x)
\end{multline*}

\begin{multline*}
J(\theta_{1})
\end{multline*}

Tarkime, kad tai mūsų duomenų rinkinys, turi tik trys taškus kurie yra (1, 1), (2, 2), (3, 3), o mes pasirinkome <code>**𝜃1 reikšmę, lygią 1**</code>. Hipotezė čia atrodys kaip ši tiesi linija.

\begin{multline*}
h_{\theta} = \theta_{1}x + \theta_{0}
\end{multline*}

\begin{multline*}
h_{\theta} = 1x + 0
\end{multline*}

Tačiau norime išsiaiškinti, kokia reikšmę turi J(θ1), kai <code>**θ1 yra lygus 1**</code>

$$J(\theta_{1}) = \dfrac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)} ) -y^{(i)})^2 $$
$$  = \dfrac{1}{2m}\sum_{i=1}^{m}(\theta_{1} x^{i} - y^{i})^2 =  \dfrac{1}{2m}(0^2 + 0^2 + 0^2) = 0^2 = 0 $$

Konkrečiam duomenų rinkiniui turime tris mokymo pavyzdžius ir jei θ1 yra lygus 1. Tada paaiškėja, kad:

$$ h_{\theta}(x^{(i)}) = y^{(i)}$$

Dabar žinome kad : 
$$J(1) = 0 $$

Atsižvelgdami į šias parametrų reikšmes, norime pavaizduoti atitinkamą kainos funkciją J.
Nuo praeito karto, kai turėjome tik vieną teta parametrą ir mūsų J grafikas buvo tiesiog plokščias. Dabar turime du parametrus ir mūsų situacija tampa šiek tiek sudėtingesnė. Su vienu parametru mūsų kainos funkcija turėjo tokią galima sakyti dubens formą (parabolę), o dabar su dviem parametrais J funkcija bus panaši į dubenėlio formą <code>**convex**</code>. Ir iš tikrųjų, priklausomai nuo mūsų duomenų rinkinio, galime gauti įvairių formų kainos funkciją.

Turbūt jau šiek tiek suprantame, kaip kaštų funkcijos J reikšmės atitinka skirtingas hipotezes ir kaip geresnė hipotezė gali atitikti taškus, kurie yra arčiau šios kaštų funkcijos J minimalios reikšmes. Taigi iš tikrųjų mes norime turėti <code>**efektyvų algoritmą, leidžianti automatiškai rasti theta parametrų vertes, kuri sumažina kainos funkciją J.**</code> Tas algoritmas bus <code>**Gradient descent**</code> , gradientinio nusileidimo algoritmas.

<code>**Tai algoritmas, kurį naudosime siekdami sumažinti savo kainos funkciją J**</code>. Tiesą sakant, tai plačiai naudojamas optimizacijos algoritmas,jis nėra naudojamas tik tiesinei regresijai. <code>**Oficialiau šis algoritmas fiksuoja visas dalines kelių kintamųjų funkcijos išvestines**</code>

Jis naudojamas visoje mašininio mokymosi srityje ir gali būti naudojamas siekiant sumažinti kitas funkcijas, ne tik kainos funkciją. Gradiento nusileidimas yra visur paplitęs neuroninių tinklų ir kitų mašininio mokymosi modelių mokyme. Kiekvieną kartą, kai turite modelį, kurio parametrai treniruočių metu atnaujinami pagal kainos funkciją, galite būti beveik tikri, kad po gaubtu yra šis algoritmas.
