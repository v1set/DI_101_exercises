## #5

Kokias operacija reikia atlikti su kvadratinę matrica D kad rezultatas būtų taip kaip parodyta žemiau ?

\begin{multline*}
D = 
\begin{bmatrix}
3 & 1.5 & 2 \\
-1 & 1 & 0.5 \\
-0.25 & 0.2 & 0.25 \\
\end{bmatrix}
\
\rightarrow
\
\begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}
\end{multline*}

<code>**Pastaba**</code> atliekant skaičiavimus su numpy jūsų rezultatų reikšmės matricoje gali pavirsti į pavyzdžiui tokias reikšmes: <code>**1.12757026e-17, -2.77555756e-17, 5.20417043e-18**</code>, dėl to nesijaudinkite tai yra labai maži skaičiai, tai indikuoja <code>**e-17, e-18 ir pan.**</code> esantis gali. Šiuos skaičius suapvalinus gautume 0.

```python
# TODO: sprendimą rašykite čia
D = np.array([
    [ 3   , 1.5, 2   ],
    [-1   , 1  , 0.5 ],
    [-0.25, 0.2, 0.25]
])

required_action = np.linalg.inv(D)

result = D.dot(required_action)
print(result)
```

## Išvestinės

## #6

Kokia yra šios funkcijos išvestinė ? : <br>
\begin{multline*}
f(x) = 2x^2
\end{multline*}

```python
#TODO:
# jūsų atsakymas turi būti čia:
from scipy.misc import derivative

def formula(x):
    return 2*x**2

def deriv(x):
    return derivative(formula, x)

y = np.linspace(-2.5, 5)
plt.plot(y, deriv(y), color='green', label='Išvestinė')
print(deriv(y))
print(y, deriv(y))
plt.plot(y, formula(y), color='black', lw=3 ,label='f(x)=2x²')
plt.scatter([0,2], [formula(0),formula(2)], s=100, c='r')

plt.legend();
```

## #7

Kokia yra šios funkcijos išvestinė ? : <br>

\begin{multline*}
f(x) = 2 + x^4
\end{multline*}

```python
#TODO:
# jūsų atsakymas turi būti čia:
from scipy.misc import derivative

def formula(x):
    return 2+x**4

def deriv(x):
    return derivative(formula, x)

y = np.linspace(-2.5, 5)
plt.plot(y, formula(y), color='black', lw=3 ,label='f(x)=2+x^4')
plt.plot(y, deriv(y), color='green', label='Išvestinė')
print(deriv(y))
print(y, deriv(y))
plt.scatter([0,2], [formula(0),formula(2)], s=100, c='r')

plt.legend();
```

## #9

Turime duomenų rinkinį ir ne ityn gerą modelį.

$$h_{\theta}(x) = 10 + 1x$$

```python
# %load ../scripts/_linear_regression_sample_c.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston

plt.figure(figsize=(12,8))

data_set = load_boston()
x = data_set.data[:,5]
y = data_set.target

x = x[:10]
y = y[:10]

theta_0 = -100
theta_1 = 20

y_hat =  theta_0 + theta_1 * x

plt.plot(x, y_hat, '-y')
plt.scatter(x, y, s=80)

plt.xlabel("kambariai")
plt.ylabel("kaina tūkst.")
plt.grid(True)
```

Šio duomenų rinkinio reikšmės yra:

```python
x = [6.575, 6.421, 7.185, 6.998, 7.147, 6.43,  6.012, 6.172, 5.631, 6.004]
y = [24,  21.6, 34.7, 33.4, 36.2, 28.7, 22.9, 27.1, 16.5, 18.9]
```

Taikydami kainos funkciją <code>**(vidutinę kvadratinę klaidą, angl. mean squared error)**</code>, apskaičiuokite modelio su koeficientais 𝜃0 = 10, 𝜃1 = 1, klaidos kainą:

$$J(\theta_{1}) = \dfrac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^{(i)} ) -y^{(i)})^2 $$
$$h_{\theta}(x) = 10 + 1x$$

```python
x = np.array(x)
y = np.array(y)
#TODO apskaičuotą reikšmę įrašykite čia vietoje 0:
cost = 0


# h_theta = 10 + 1*x

def quad_sum():
    m = x.size
    return ( 1 / (2*m) ) * sum(  [ i**2 for i in ((10 + 1*x) * (x - y)) ] )

cost = quad_sum()
print(cost)

```
