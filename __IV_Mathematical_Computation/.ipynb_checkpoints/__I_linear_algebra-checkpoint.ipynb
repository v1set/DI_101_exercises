## Matrica

Trumpai apžvelkime, kas yra matricos ir vektoriai. Tačiau rekomenduoju pasižiurėti labai gerai sudarytą kursą apie <code>**[linijinę algebrą, autorius 3Blue1Brown](https://www.youtube.com/playlist?list=PLZHQObOWTQDPD3MizzM2xVFitgF8hE_ab)**</code>
, glausta informacija apie [matrica lietuviškai][https://lt.wikipedia.org/wiki/Matrica_(matematika)]. <code>**Matrica**</code> galime laikyti stačiakampiu skaičių masyvu, paprastai rašomu tarp laužtinių skliaustų. Žemiau yra dvi matricos su keletą skaičių, matricso formoje gali būti jūsų duomenys, pvz., mašininio mokymosi duomenų savybės, arba duomenys iš kažkur kitur, konkrečios reikšmės šiuo metu nėra svarbios.

\begin{align}
\begin{bmatrix}
1500 & 544 \\
1241 & 945 \\
212 & 678 \\
185 & 211
\end{bmatrix} 
\begin{bmatrix}
1 & 2 & 3\\
4 & 5 & 6
\end{bmatrix} 
\end{align}

Tai tik dar vienas būdas pasakyti, kad tai 2D (dvimatis) masyvas. Kita ypatybė, kurios mums reikia, yra ta, kad matricos matmenys bus parašyti kaip <code>**eilučių skaičius**</code>, padaugintas iš <code>**stulpelių skaičiaus**</code>, tai matricos dimensijos/matmenys.
<code>**Matricos matmenys: eilučių skaičius x stulpelių skaičius.**</code> Pirmame pavyzdyje viršuje kairėje yra 4 eilutės ir 2 stulpeliai, todėl tai <code>**4x2 matrica**</code> . Antrasis pavyzdys yra <code>**2x3 matrica**</code>.<br>
Dažnai matysite, kad matrica parašyta taip <code>**(m yra eilutės ir n yra stulpeliai)**</code>:

$$
\mathbb{R}^{m x n}
$$

### Matricos elementai (matricos įrašai)

Matica visada yra žymima <code>**didžiąja raide**</code>. Bet kaip nurodyti konkretų matricos elementą ? :

\begin{align}
A =
\begin{bmatrix}
111 & 544 \\
1241 & 945 \\
212 & 678 \\
185 & 211
\end{bmatrix} 
\end{align}

<code>**i, j žymės eilučių ir kolonų indeksus**</code><br>
<code>**Pastaba:**</code>(matematikoje dažniausiai matricos ir vektorių indeksavimas prasideda nuo 1, skirtingai nuo programavimo kur indeksuojame masyvų narius nuo 0)
<br>

\begin{align}
A_{ij} = \text{i,j elementas yra} \ i^{\text{th}} \text{eilutejė,} \ \ j^{\text{th}} \text{kolonoje};
\end{align}

\begin{align}
A_{11} = 111, A_{12} = 544, A_{32} = 678
\end{align}


## Vektorius

Vektorius yra "ypatingas" matricos atvejis. Iš esmės vektorius yra matrica, turinti tik 1 stulpelį, todėl mes turime
<code>**nx1**</code> matrica. Matricą su vienu stulpeliu vadiname vektoriumi, žymima <code>**mažosiomis raidėmis**</code>.

\begin{equation*}
y = 
\begin{bmatrix}
112 \\
445 \\
524 \\
258
\end{bmatrix}
\in \mathbb{R}^{n}
\end{equation*}

Nurodome vektoriaus elementus standartine žyma
\begin{align}
y_{i} = i^{\text{th}} \text{elementas};
\end{align}

\begin{align}
y_{1} = 112, y_{2} = 445, y_{3} = 524
\end{align}

Tiesinėje algebroje yra dvi indeksavimo sutartys <code>**1 indeksuotė ir 0 indeksuotė**</code>,tai taip pat atitinka kai kurias programavimo kalbas. <code>**Matematikoje indeksuotė nuo 1 yra labiau paplitusi**</code>, tačiau mašininio mokymosi programose <code>**indeksuotė nuo 0**</code> suteikia daugiau techninio patogumo. <code>**Šiame kurse matematiniu koncepcijų aptarime dažniausiai naudosiu indeksuotę nuo 1**</code> ir aiškiai pasakysiu, kada turėsime pereiti prie indeksavinuo nuo 0.

## Matricos daugyba su matrica

Galime padauginti dvi matricas kartu. Šie metodai naudojamas sprendžiant mašininio mokymosi problemas, susijusias su parametrų optimizavimu tarkime tiesinėje regresijoje, šią optimizacijos<code>**galima atlikti vienu kartu, ir nereikalauja pasikartojančio iteratyvaus algoritmo naudojimo, pvz., gradiento nusileidimo algoritmas, [gradient descent](https://en.wikipedia.org/wiki/Gradient_descent)**</code>. Kai kalbėsimės apie šį algoritmą, paaiškės, kad matrica – matricos daugyba yra vienas iš pagrindinių dalykų kurį turime žinoti.<br>
Pakartokite klasės praktikos pratimą:

\begin{align}
\begin{bmatrix}
1 & 2 & 2 & 2\\
0 & 1 & 0 & 0\\
2 & 0 & 0 & 4
\end{bmatrix} 
\times
\begin{bmatrix}
1 & 2 \\
3 & 0 \\
1 & 1 \\
4 & 2
\end{bmatrix} 
=
\end{align}

Padarykime tai etapais, ištraukdami kai kuriuos stulpelius, tvarkykime šią problemą tai kaip paprasčiausią matricos ir vektoriaus dauginimą tik žignsniais.

\begin{align}
\begin{bmatrix}
1 & 2 & 2 & 2\\
0 & 1 & 0 & 0\\
2 & 0 & 0 & 4
\end{bmatrix} 
\times
\begin{bmatrix}
1 \\
3 \\
1 \\
4 
\end{bmatrix} 
=
\begin{bmatrix}
17 \\
3 \\
18
\end{bmatrix} 
\end{align}


```python
import numpy as np

A  = np.array([
    [1, 2, 2, 2],
    [0, 1, 0, 0],
    [2, 0, 0, 4]
])

c = np.array([1, 3, 1, 4])

result = A.dot(c)
result
```

Atlikite tą patį su kitu antrosios matricos stulpeliu.
\begin{align}
\begin{bmatrix}

\begin{align}
\begin{bmatrix}

\end{bmatrix} 
\times
\begin{bmatrix}

\end{bmatrix} 
=
\begin{bmatrix}

Tai yra paprastas principas, kaip padauginti matricą iš kitos matricos. Bet iš tikrųjų programiškai tai galite padaryti ir vienu kartu.

```python
A  = np.array([
    [1, 2, 2, 2],
    [0, 1, 0, 0],
    [2, 0, 0, 4]
])
B = np.array([
    [1, 2],
    [3, 0],
    [1, 1],
    [4, 2]
])
result = A.dot(B)
result
```

\begin{align}

\end{align}

### Praktinis taikymo pavyzdys

Tai vadinu <code>**vieno šūvio**</code> metodu.
Tai labai efektyvus ir supaprastintas būdas spręsti mašininio mokymosi problemas. Mes plačiau kalbėsis apie šį metodą <code>**vektorizavimo**</code> temose.
Tai puikus matematinis triukas. Turime kai kurių namų dydžių pavyzdžius ir hipotezę, kuri numato būsto kainą. Ir mes norime apskaičiuoti h(x) kiekvienam namui. Tai galime padaryti efektyviai skaičiavimo būdu, naudodami <code>**matricos vektorių dauginimą**</code>

<code>**dydžiai**</code>:
<code>50</code>
<code>77</code>
<code>52</code>
<code>218</code>


$$ h_{\theta}(x) = -0.1 + 5x$$

Leidžiame sudaryti mūsų duomenų imčių matricą ir hipotezės parametrų (koeficientų) vektorių ir padauginti juos kartu.

\begin{equation*}
\begin{bmatrix}
-0.1 \\
5
\end{bmatrix}^{\top}
\times
\begin{bmatrix}
1 & 50 \\
1 & 77 \\
1 & 52 \\
1 & 218
\end{bmatrix}
\end{equation*}

Kurdami tokią programa naudojame turimus duomenų pavyzdžius ir norime panaudoti savo hipotezę kainoms prognozuoti,<code>** arba kitap nustatyti y reiškmė.**</code> Tai galime padaryti vos per kelias kodo eilutes, mums tereikia tinkamos bibliotekos, kad galėtume atlikti tiesinės algebros operacijas, pvz., matricos dauginimą. Kaip alternatyvą tai patgalime tai padaryti naudodami ciklus ir kitas operacijos tačiau tai bus daug sudėtingiau ir užtruks ilgiau suprogramuoti. Naudojant linijinę algebrą įvairiose programavimo kalbose, tai leidžia ne tik paprasčiausiai koduoti, bet ir dėl subtilių priežasčių kompiuterinio skaičiavimo požiūriu daug efektyviau alikti verčių prognozes.


<code>**Hypothesis**</code>:
\begin{multline*}
h_{\theta}(x)=  \theta_{0}x_{0} + \theta_{1}x_{1} 
\end{multline*}

\begin{multline*}
h_{\theta}(x)= \theta^\top x
\end{multline*}

<code>**savybės**</code>:
<code>50</code>
<code>77</code>
<code>52</code>
<code>218</code>

<code>**Keletas konkuruojančių hipotezių**</code>:

\begin{multline*}
h_{\theta}(x) = -0.1 + 5.25x
\end{multline*}
\begin{multline*}
h_{\theta}(x) = 55 + 1.1x
\end{multline*}
\begin{multline*}
h_{\theta}(x) = -20 + 1.4x
\end{multline*}

Mes galime pritaikyti visas tris galimas hipotezes visoms savybėms ir tai galime padaryti labai efektyviai naudodami matricos - matricos dauginimą ir gauti tris prognozes.

\begin{align}
\begin{bmatrix}
1 & 50 \\
1 & 77 \\
1 & 52 \\
1 & 218
\end{bmatrix}
\times
\begin{bmatrix}
-0.1 & 55  & -20\\
5.25 & 1.1 & 1.4
\end{bmatrix}
=
\begin{bmatrix}
262.4 & 110 & 50   \\
404.15 & 139.7 & 87.8 \\
272.9 & 112.2 & 52.8 \\
1144.4 & 294.8 & 285.2 
\end{bmatrix}
\end{align}

```python
import numpy as np

X = np.array([
    [50],
    [77],
    [52],
    [218]
])

## ypač svarbu atlikti .reshape() kad suformuoti vektorių
θ = np.array([-0.1, 5.25]).reshape(-1,1)

##############
print("X matrix savybės/reikšmės")
print(type(X))
print(X.ndim)
print(X.shape)
##############
print("θ vector savybės/reikšmės")
print(type(θ))
print(θ.ndim)
print(θ.shape)
```

```python
X = np.insert(X, 0, 1, axis=1)
```

\begin{multline*}

\begin{bmatrix}

77 \\
52 \\
218
\end{bmatrix}
\rightarrow
\begin{bmatrix}
1 & 50 \\
1 & 77 \\

\end{bmatrix}
\end{multline*}

```python
X.shape, θ.shape
```

```python
y = X.dot(θ)
y
```

## Matricos daugybos savybės

Matricos daugyba yra labai naudinga, nes galime sutalpinti daug skaičiavimo į vieną duomenų struktūros operaciją. Yra keletas šios operacijos savybių, kurias turėtume žinoti ir būti atidūs, kai naudojate šį metodą.

### Komutacinė savybė 

<code>**Matricos daugyba NĖRA komutacinė**</code><br>
Dirbant su realiaisiais skaičiais , skaliarais, daugyba yra komutacinė ir tvarka nesvarbu.

\begin{multline*}
3 \times 2 = 2 \times 3
\end{multline*}

<code>**Tačiau tai netinka matricos daugybai**</code><br>
Tegu A ir B yra matricos

\begin{multline*}
A \times B \neq B \times A
\end{multline*}

Negalima  pakeisti matricų dauginimo tvarkos.<br>
Jei sukeisime tvarką tarp šių dviejų matricų, gausime skirtingus sprendimus.

\begin{multline*}
\begin{bmatrix}
1 & 1 \\
0 & 0
\end{bmatrix}
\times
\begin{bmatrix}
0 & 0 \\
2 & 0
\end{bmatrix}
=
\begin{bmatrix}
2 & 0 \\
0 & 0
\end{bmatrix}
\end{multline*}


\begin{multline*}
\begin{bmatrix}
0 & 0 \\
2 & 0
\end{bmatrix}
\times
\begin{bmatrix}
1 & 1 \\
0 & 0
\end{bmatrix}
=
\begin{bmatrix}
0 & 0 \\
2 & 2
\end{bmatrix}
\end{multline*}

Apskritai matricos daugybos tvarkos pakeitimas gali netgi pakeisti rezultato matmenis.
Jei <code>**A x B**</code> yra <code>**mxn ir nxm**</code>, tada <code>**A x B**</code> bus <code>**mxm**</code> matmenų struktūros matrica ir <code>**B x A**</code> bus <code>**nxn**</code> matmenų struktūra.

### Asociatyvi savybė

Nagrinėdami skaliarus (realius skaičius), turime realiųjų skaičių daugyb asociatyvines savybes.

\begin{multline*}
3 \times 2 \times 2 = (3 \times 2) \times 2 = 3 \times (2 \times 2) = 12
\end{multline*}

<code>**Matricų sandauga yra asociatyvi**</code><br>
\begin{multline*}
A \times B \times C = A \times (B \times C) = (A \times B) \times C
\end{multline*}

\begin{multline*}
D = B \times C, A \times D
\end{multline*}


\begin{multline*}
E = A \times B, E \times C
\end{multline*}

Abi šios parinktys suteiks mums tą patį sprendimą.

## Tapatybės matrica

Tarkime, kad bet kurį realųjį skaičių (a) padauginame iš 1.

\begin{multline*}
1 \times a = a \times 1 = a
\end{multline*}

Tai tapatybės operacija ir ji turi atitikti šią lygtį.
Tiesinėje algebroje, susijusioje su matricomis, yra <code>**tapatybės matrica**</code>, pažymėta <code>**I**</code> ir tai reiškia <code>**kvadratinę matricą**</code> jos matmenys yra <code>**nxn**</code>.

```python
import numpy as np

A = np.eye(8)
A
```

Jis turi savybę <code>**1 išilgai įstrižainėje ir 0 visur kitur**</code>.

Bet kuriai matricai A

\begin{multline*}
A \times I = I \times A = A
\end{multline*}

## Inversija ir transpozicija

Kaip atvirkštinis ryšys su realiais skaičiais. Atminkite, kad 1 skaičius realiųjų skaičių erdvėje atlieka <code>**tapatybės**</code> vaidmenį, nes akivaizdu, kad padauginimas iš bet ko yra lygus jam pačiam. Ir yra savybė, kad kiekvienas realusis skaičius turi <code>**atvirkštinią reikšmę**</code>, pavyzdžiui, 3·(3ⁿ) = 1 , kur n=-1. (3ⁿ) kur n=-1 yra toks pat kaip 1/3. Tai gali būti bet koks realusis skaičius, produktas visada bus tapatybė.<br>

```python
3 * (3**-1)
```

<code>**Tačiau tikrojo realujų erdvėje ne viskas turi atvirkštinę reikšmę**</code>, pvz., 0, jo atvirkštinė vertė 1/0 būtų neapibrėžta.
Taigi, kaip tai taikoma matricai ir kaip kompiuteriu apversti matricą. Idėja yra tokia:


\begin{multline*}
AA^{-1}= A^{-1}A = I
\end{multline*}

Jei A yra <code>**mxm matrica, kitaip vadinama kvadratine matrica**</code> ir jei ji turi atvirkštinę formą, tada jos sandauga yra lygi tapatybės matricai.


\begin{multline*}
\begin{bmatrix}
3 & 4 \\
2 & 16
\end{bmatrix}^{-1}
 = 
\dfrac{1}{3 \times 16 - 4 \times 2 }
\times
\begin{bmatrix}
16 & -4 \\
-2 & 3
\end{bmatrix}  
= 0.025 \times
\begin{bmatrix}
16 & -4 \\
-2 & 3
\end{bmatrix}  
=
\begin{bmatrix}
0.4 & -0.1 \\
-0.05 & 0.075
\end{bmatrix}  
\end{multline*}

Tačiau šiais laikais retai kas tai daro rankomis, dauguma programinės įrangos programavimo kalbų linijinės algebros bibliotekos turi įmontuotus įrankius šiai operacijai atlikti.

```python
import numpy as np

A  = np.array([
    [3, 4 ],
    [2, 16]
])
A
```

```python
B = np.linalg.inv(A)
B
```

```python
type(A)
```

## Tenzoriai

<code>**Scalar (rank-0 tensor)**</code>

\begin{multline*}
\mathbf{x} \in \mathbb{R}
\end{multline*}

<code>**Vector (column) (rank-1 tensor)**</code>

\begin{multline*}
\mathbf{x} \in \mathbb{R}^{n \times 1}
\end{multline*}

<code>**Vector (row) (rank-1 tensor)**</code>

\begin{multline*}
\mathbf{x}^{\top} \in \mathbb{R}^{1 \times n}
\end{multline*}

<code>**Matrix (rank-2 tensor)**</code>

\begin{multline*}
\mathbf{X} \in \mathbb{R}^{m \times n}
\end{multline*}

<code>**Tensor (rank-3 tensor)**</code>

\begin{multline*}
\mathbf{X} \in \mathbb{R}^{m \times n \times p}
\end{multline*}

Dažnai naudosime 𝐗 simbolį kaip specialų susitarimą, nurodydami kad tai <code>**dizaino matricą**</code>. Tai matrica, kurioje yra mokymo pavyzdžiai ir savybės (įvestys) ir struktūra bus<code>**𝐗 ∈ ℝ 𝑛×𝑚**</code>, nes dažnai naudojamas <code>**𝑛**</code> atkreipti dėmesį į pavyzdžių skaičių literatūroje daugelyje disciplinų.

\begin{multline*}
\mathbf{X} = 
\begin{bmatrix}
x^{[1]}_{1} & x^{[1]}_{2} & \ ... \ & x^{[1]}_{m} \\
x^{[2]}_{1} & x^{[2]}_{2} & \ ... \ & x^{[2]}_{m} \\
. & . & \ ... \ & . \\
x^{[n]}_{1} & x^{[n]}_{2} & \ ... \ & x^{[n]}_{m} \\
\end{bmatrix}
\end{multline*}

<br>

<code>**2'a savybės reikšmė iš 1'ojo mokymų pavyzdžio:**</code>

\begin{multline*}
x^{[1]}_{2}
\end{multline*}

\mathbf{X} = 
\begin{bmatrix}

x^{[n]}_{1} & x^{[n]}_{2} & \ ... \ & x^{[n]}_{m} \\

\end{multline*}

Kompiuteris iš esmės gali panaudoti žemesnio lygio procedūras, kad lygiagrečiai apskaičiuotų tam tikrus dalykus.

   1. elementų dauginimas taškinei sandaugai apskaičiuoti

### Kelių bandymų pavyzdžių išvesties apskaičiavimas vienu metu

Dabar vietoj funkcijų vektoriaus turime šią dizaino matricą ir w, kuri yra:
\begin{multline*}
\mathbf{X} \in \mathbb{R}^{n \times m}
\end{multline*}


\begin{multline*}
\mathbf{w} \in \mathbb{R}^{m}
\end{multline*}

Štai kodėl 𝐰 yra ne vektorius, o 𝑚 × 1 matrica

\begin{multline*}
\mathbf{X}\mathbf{w} + b = \mathbf{z}
\end{multline*}

Taigi, padauginus 𝐗 ir 𝐰 sandauga būtų ℝ𝑛×1. Štai kodėl mes turime šiuos papildomus vektorių matmenis, nes linijinėje algebroje tradicine prasme negalime apskaičiuoti/dauginti matricos ir vektorių daugybos, galime padaryti tik <code>**taškų sandaugą**</code> tarp dviejų vektorių. ir <code>**matricos daugyba**</code> tarp dviejų matricų.

kur

\begin{multline*}
\mathbf{X} = 
\begin{bmatrix}
x^{[1]}_{1} & x^{[1]}_{2} & \ ... \ & x^{[1]}_{m} \\
x^{[2]}_{1} & x^{[2]}_{2} & \ ... \ & x^{[2]}_{m} \\
. & . & \ ... \ & . \\
x^{[n]}_{1} & x^{[n]}_{2} & \ ... \ & x^{[n]}_{m} \\
\end{bmatrix},
\mathbf{w}=
\begin{bmatrix}
w_{1} \\
w_{2} \\
.     \\
w_{m}
\end{bmatrix}
\end{multline*}

\begin{multline*}
\mathbf{z}=
\begin{bmatrix}
\mathbf{w}^{\top}\mathbf{x}^{[1]} +b \\
\mathbf{w}^{\top}\mathbf{x}^{[2]} +b \\
. \\
\mathbf{w}^{\top}\mathbf{x}^{[n]} +b 
\end{bmatrix}
=
\begin{bmatrix}
z^{[1]} \\
z^{[2]} \\
. \\
z^{[n]} 
\end{bmatrix}
\end{multline*}

Tiek `NumPy`, tiek `PyTorch` nėra labai išrankūs tradiciniam tiesinės algebros žymėjimui.<br>
`view(2,3)` sukuria 2 x 3 matricą su 6 elementais nuo 0 iki 5.

```python
X = torch.arange(6).view(2,3)
```

```python
X
```

Čia galime atsisakyti antro matmens
\begin{multline*}
\mathbf{w} \in \mathbb{R}^{m \ \times \ \ \hspace-2.2ex\diagup 1 }
\end{multline*}

```python
w = torch.tensor([1, 2, 3])
```

taigi 𝐰 yra 1 matmens vektorius ir mes galime padauginti 𝐗 iš 𝐰

```python
X.matmul(w)
```

Ir tai suteiks mums vektorių su reikšmėmis [8, 26]. Taigi tai taip pat vektorius su 1 matmeniu ir dviem elementais. Tačiau prie šio vektoriaus galime pridėti kitą dimensiją naudodami funkciją `view(-1,1)`, kuri yra tokia pati kaip `NumPy` `reshape`.

```python
w = w.view(-1, 1)
```

```python
m
```

dabar 𝐰 yra dviejų matmenų <code>**stulpelio vektorius**</code>, todėl tai yra matrica.


\begin{multline*}
\mathbf{w}=
\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix}
\in \mathbb{R}^{m \times 1}
\end{multline*}

<code>**Pastaba:**</code> ši operacija nebūtina. Bet matome, kad išvestis yra 𝑛 × 1 matmenys, kur ankstesnė išvestis buvo tik 𝑛 elementai (1 ašis)

```python
X.matmul(w)
```

Kadangi terminas taip pat negalioja tradicinėje tiesinėje algebroje, mes negalime pridėti skaliarinės reikšmės prie vektoriaus

\begin{multline*}
\mathbf{X}\mathbf{w} + b = \mathbf{z}
\end{multline*}

Tradicinėje tiesinėje algebroje taip turėtų būti

\begin{multline*}
\mathbf{X}\mathbf{w} + 1_{m}b = \mathbf{z}
\end{multline*}

Tačiau mašininio mokymosi inžinieriai yra tiesiog tingūs žmonės, ir mes to nedarome, nes tai būtų per daug pastangų.
