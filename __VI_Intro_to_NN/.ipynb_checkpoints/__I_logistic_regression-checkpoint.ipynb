Klasifikavimo problema, kai kintamasis <code>**y**</code>, kurį norime numatyti yra <code>**diskreti reikšmė**</code>. Prieš tai mūsų priklausomas kintamasis buvo <code>**tęstinis**</code>
Mes aptarsime algoritmą, kuris vadinasi logistine regresija. Tai vienas populiariausių ir plačiausiai naudojamų mokymosi algoritmų.<br>

<code>**Yra keletas klasifikavimo algoritmų:**</code>
 * Logistic regression
 * Naïve Bayes
 * K-Nearest Neighbours
 * Decision Tree
 * Random Forest
 * Support Vector Machine

<code>**Kai kurios klasifikavimo problemos**</code>:
  * Medicinos: navikas yra piktybinis ar ne
  * Finansai: sandoriai yra apgaulingi, ar ne
  * El. paštas: ar šis el. laiškas yra šlamštas, ar ne

Bandome numatyti kintamąjį y:<br>

\begin{multline*}
y \in (0,1)
\end{multline*}

 * 0: <code>**neigiama klasė**</code>: (t.y teisėtas sandoris)
 * 1: <code>**positive class**</code>: (t.y apgaulinga operacija)

\begin{multline*}
y \in (0, 1, 2, 3, 4)
\end{multline*}

\begin{multline*}

Kas iš tikrųjų atsitiks, jei klasifikavimo problemai naudosite linijinę regresiją. Tokio tipo problemos atveju žinome, kad <code>**y yra 0 arba 1**</code>, bet jei naudojama tiesinė regresija, hipotezė gali <code>**išvesti reikšmę, kuri yra didesnė už 1 arba mažesnė už 0**</code> net jei visi mokymo pavyzdžiai turi etiketes, kurios yra lygios 0 arba 1.

<code>**Klasifikacija**</code>: y = 0 or 1 <br>

\begin{multline*}
h_{\theta}(x) \ \text{can be} \ > 1 \ \text{or} \ <0
\end{multline*} 

Taigi turime panaudoti algoritmą, kuris formaliai vadinamas logistine regresija ir turi savybę, kad <code>**prognozuojama išvestis visada yra nuo 0 iki 1**</code>

<code>**Logistinės regresijos išvestis**</code>:<br>

\begin{multline*}
0 \leq h_{\theta}(x) \leq 1
\end{multline*} 

## Logistinės regresijos apibrėžimas

<code>**Logistinės regresijos modelio (hipotezės) išvestis:**</code>

\begin{multline*}
0 \leq h_{\theta}(x) \leq 1
\end{multline*}

<code>**Tiesinės regresijos modelis (hipotezė):**</code>

\begin{multline*}
h_{\theta}(x) = \theta^{\top}x
\end{multline*}

<code>**[Sigmoidinė (logistinė) funkcija](https://en.wikipedia.org/wiki/Sigmoid_function#:~:text=4%20Applications-,Definition,refer%20to%20the%20same%20object.)**</code><br> <code>**Pastaba: Statistikoje tai vadinama logit arba  log-odds funkcija.**</code>

\begin{multline*}
h_{\theta}(x) = g(\theta^{\top}x)
\end{multline*}


\begin{multline*}
g(z) = \dfrac{1}{1+e^{-z}}
\end{multline*}

<code>**Logistinės regresijos modelis (hipotezė):**</code>
\begin{multline*}
h_{\theta}(x) = \dfrac{1}{1+e^{-\theta^{\top}x}}
\end{multline*}

<code>**Kas yra e , Eulerio skaičius ?**</code>, tai matematinė konstanta, maždaug lygi <code>**2.71828**</code>, ir gali būti apibūdinamas įvairiais būdais. Tai yra natūralaus logaritmo pagrindas<code>**[apie tai daugiau informacijos rasite čia](https://www.youtube.com/watch?v=pDFcu_wLOzo) ir [čia](https://www.youtube.com/watch?v=m2MIpDrF7Es)** </code>

<code>**Eulerio skaičiaus savybės:**</code>
<br>
\begin{multline*}
e = \lim\limits_{n \to \infty}( 1 + \dfrac{1}{n})^n
\end{multline*}

\begin{multline*}
e = \lim\limits_{n \to 0}( 1 + n)^{\dfrac{1}{n}}
\end{multline*}


\begin{multline*}
e = \sum_{n=0}^{\infty} \dfrac{1}{n!} = 1 + \dfrac{1}{1} + \dfrac{1}{1 \times 2} + \dfrac{1}{1 \times 2 \times 3} + ...
\end{multline*}

\begin{multline*}
y = e^x = \dfrac{dy}{dx}
\end{multline*}
<br>

<code>**Sigmoidinė (logistinė) funkcija**</code>

\begin{multline*}
g(\theta^{\top}x) = \dfrac{1}{1+e^{-\theta^{\top}x}}
\end{multline*}

## Hipotezės išvestis

Logistinės regresijos hipotezės išvestį interpretuokite kaip tikimybę.

\begin{multline*}
h_{\theta}(x) = \ \text{estimated probabilty that y} \ = 1 \ \text{on input} \ x 
\end{multline*}

<br>

Lets say we have feature x vector that has x_0 that is always equal to 1 and x_1 that is test cell size.
Tarkime, kad turime savybių vektorių x , kuris turi x_0, kas yra  visada yra lygus 1, ir x_1, kuris yra lastelės diametro dydis.

\begin{multline*}
x = 
\begin{bmatrix}
x_{0} \\
x_{1 }
\end{bmatrix}
 =
\begin{bmatrix}
1 \\
cellsize
\end{bmatrix}
\end{multline*}

<br>
Mūsų hipotezė rodo, kad:

\begin{multline*}
h_{\theta}(x) = 0.85
\end{multline*}

Tai rodo, kad šis mėginys turi 85% tikimybę, kad audinys yra piktybinis.<br>
<code>**Tikimybė kad y = 1, su duotuoju x kuri yra parametrizuotas su θ**</code>

\begin{multline*}
h_{\theta}(x) = P(y= 1\vert x ; \theta)
\end{multline*}
Taigi remsimės savo hipoteze, kuri pateiks mums <code>**tikimybės, kad y = 1** įverčius</code> ir kadangi tai yra klasifikavimo problema, <br>žinome, kad galimos y reikšmės = 0 arba 1

## Kainos funckija

<code>**Kaip automatiškai pasirinkti parametrus θ**</code>,kad atsižvelgiant į duomenų rinkinį galėtume automatiškai optimizuoti parametrus prie mūsų duomenų. <br>
Apibrėžkime optimizavimo tikslą ir <code>**kainos funkciją, kurią naudosime parametrams pritaikyti.**</code><br>

Prižiūrimo mokymosi kategorijos logistinės regresijos modelio pritaikymas:

<code>**Turime m duomenų pavyzdžių rinkinį, kuris atrodo taip:**</code><br>

\begin{multline*}
{(x^{(1)}, y^{(1)}), (x^{(2)}, y^{(2)}), ..., (x^{(m)}, y^{(m)}) }
\end{multline*}

<kodas>**Mūsų m pavyzdžių savybių vektorius kurio matmuo yra n+1. x_0 yra lygus 1 ir kadangi tai yra klasifikavimo problema, mūsų duomenų rinkinio įrašai turi savybę, kad kiekviena etiketė t.y y yra 0 arba 1:**</code><br>

\begin{multline*}
x \in 
\begin{bmatrix}
x_0 \\
x_1 \\
x_2 \\
... \\
x_n
\end{bmatrix}
\in\mathbb{R^{n+1}},
x_0 = 1, y\in {0,1}
\end{multline*}

<code>**Hipotezė:**</code><br>

\begin{multline*}
h_{\theta}(x) = \dfrac{1}{1+e^{-\theta^{\top}x}}
\end{multline*}

<code>**Kaip pasirinkti/optimizuoti θ parametrus ?**</code><br>

Kai dirbome su tiesinės regresijos modeliuwe naudojome šią kainos funkcija (šiek tiek kitokia, bet apskritai principas tas pats)

<code>**Tiesinės regresijos kainos funkcija :**</code><br>

\begin{multline*}
J(\theta) = \dfrac{1}{m}\sum_{i=1}^{m}\dfrac{1}{2}(h_{\theta}(x^{(i)} ) -y^{(i)})^2
\end{multline*}

<code>**Supaprastinta vieno pavyzdžio kainos funkcija:**</code><br>

\begin{multline*}
\text{cost} (h_{\theta}(x), y) = \dfrac{1}{2}(h_{\theta}(x) -y)^2
\end{multline*}

Šios kainos funkcijos aiškinimas yra toks, kad tai yra kainos, kuriuos mes norime, kad mūsų mokymosi algoritmas sumokėtų, jei jis išveda numatytą reikšmę h𝜃(𝑥) tikrosios etiketės y atžvilgiu. Kalbant apie linijinę regresiją, jos kaina yra 1/2 karto mažesnė už kvadratinį skirtumą tarp to, ką mes prognozavome, ir faktinės vertės, kurią stebime y. Ir ši kainos funkcija puikiai veikė <code>**tiesinei regresijai**</code>, bet mus domina <code>**logistinė regresija**</code>. Jei galėtume minimizuoti šią kainos funkciją, tai veiktų. Bet tai būtų neišgaubta (non-convex) funkcija.

Tai atsitinka todėl kad<code>**logisitnės regresijos funkcija h𝜃(𝑥) turi netiesiškumą.**</code>

\begin{multline*}
h_{\theta}(x) = \dfrac{1}{1+e^{-\theta^{\top}x}}
\end{multline*}

Tai gana sudėtinga netiesinė (sigmoidinė) funkcija. O paėmus ir prijungus čia:

\begin{multline*}
\text{cost} (h_{\theta}(x), y) = \dfrac{1}{2}(h_{\theta}(x) -y)^2
\end{multline*}

Ir tada čia:

\begin{multline*}
J(\theta) = \dfrac{1}{m}\sum_{i=1}^{m}\dfrac{1}{2}(h_{\theta}(x^{(i)} ) -y^{(i)})^2
\end{multline*}

Ir tada, jei nubraižysime J(𝜃) ji turės daugybe vietinių optimumų.<code>**Jei GD algoritmas veiktų pagal tokią funkciją, jis greičiausiai net nepriartės prie globalaus minimumo:**</code> :

### Convex Išgaubtumas

Problema naudojant šią kainos funkciją yra ta, kad dėl šios labai netiesinės funkcijos, kuri atsiranda viso proceso viduryje. Taigi atrodo, kad J(𝜃) yra neišgaubta funkcija, jei apibrėžtumėte ją kaip kvadratinės kainos funkciją. Taigi turime naudoti kažką kitokio, kas būtų išgaubta ir galėtume taikyti gradiento nusileidimą.


<code>**[Convex analizė](https://en.wikipedia.org/wiki/Convex_analysis) tema nepatenka į šią paskaitą, tačiau tai gana įdomi optimizavimo teorijos sritis, kurią verta skirti laiko, kad su ja susipažintumėte.**</code>

<code>**[Vizualiai paaiškinta](https://www.youtube.com/watch?v=d0CF3d5aEGc)**</code>

<code>**Hipoteze, (log of odds)**</code><br>
\begin{multline*}
h_{\theta}(x) = \dfrac{1}{1+e^{-\theta^{\top}x}}
\end{multline*}

<code>**Kainos funkcija (logistinė regresija):**</code><br>

\begin{multline*}
\text{cost} (h_{\theta}(x), y) = \left\{ \begin{array}{rcl}
-\log(h_{\theta}(x)) \ \text{if} \ y = 1\\
-\log(1- h_{\theta}(x)) \ \text{if} \ y = 0
\end{array}\right.
\end{multline*}

<code>**if y = 1:**</code>
\begin{multline*}
-\log(h_{\theta}(x))
\end{multline*}

<code>**Mes žinome kad:**</code>
\begin{multline*}
0 \leq h_{\theta}(x) \leq 1
\end{multline*}


Vienas iš būdų tai suprasti yra grafikas <code>**log(ℎ𝜃(𝑥))**</code>, artėjant prie -∞. <code>**Mes domimės diapazonu, kai ši funkcija yra nuo 0 iki 1**</code>.<br>

<h2>Logistinės regresijos kainos funkcijos: įdomių ir pageidaujamų savybių</h2>

<code>**-log(ℎ𝜃(𝑥)) | −log(1−ℎ𝜃(𝑥))**</code><br>
    
Atkreipkite dėmesį <code>**y = 1 and ℎ𝜃(𝑥) = 1**</code> , kitais žodžiais hipiteze tiksliai prognozuoja 1 ir y yra tiksliai 1 kaip ir prognozuojama, <code>**tada kaina = 0**</code>. Ir tai pat atkreipkitė dėmėsį <code>**ℎ𝜃(𝑥) → 0, kaina → ∞**</code>. Kaina kyla į begalybę. <br>
    
_____________________
Tai užfiksuoja intuiciją, jei: <code>**ℎ𝜃(𝑥) = 0 , prognozuoja P(y = 1 | x; 𝜃) = 0, but tikrasis y = 1**</code> mokymosi algoritmas bus nubaustas labai didele kaina.<br>
</div>

<code>**if y = 0:**</code>
\begin{multline*}
-\log(1 - h_{\theta}(x))
\end{multline*}

 * <code>**Nepriklausomai nuo to, ar y = 0 or y = 1, ℎ𝜃(𝑥) = 0.5, tada kaina(ℎ𝜃(𝑥),y) > 0**</code>

## Supaprastinta kainos funkcija

Anksčiau apibrėžėme kainos funkciją <code>**vienam duomenų pavyzdžiui**</code>. Galime tai plėtoti toliau ir apibrėžti visam duomenų rinkiniui. Yra paprastesnis būdas parašyti šias lygtis, negu kad naudojamaos aukščiau, remdamiesi tuo mes pritaikysime jas gradiento nusileidimo algoritmui kad atitiktume logistinės regresijos parametrų optimizaciją.

<code>**Logistinės regresijos kainos funkcija:**</code>

\begin{multline*}
J(\theta) = \dfrac{1}{m}\sum_{i=1}^{m}Cost(h_{\theta}(x^{(i)} ),y^{(i)})
\end{multline*}

<code>**kainos funkcija vienam pavyzdžiui, kurį sukūrėme anksčiau:**</code>

\begin{multline*}
\text{Cost} (h_{\theta}(x), y) = \left\{ \begin{array}{rcl}
-\log(h_{\theta}(x)) \ \text{if} \ y = 1\\
-\log(1- h_{\theta}(x)) \ \text{if} \ y = 0
\end{array}\right.
\end{multline*}

<code>**Pastaba!: y = 0 or 1, tai yra visada!**</code>

Ir kadangi mūsų klasifikavimo uždavinyje y visada yra lygus 0 arba 1, galėsime naudoti <code>**paprastesnę kainos funkciją**</code>. Užuot įrašę šią kainos funkciją, kaip nurodyta aukščiau, dviejų atskirų eilučių masyve su abiem atskirais atvejais. Galime tai suspausti į vieną patogesnę lygtį ir tada ją panaudoti gradiento nusileidimo algoritme.

<div class="alert alert-block alert-info">
    
<code>**Paprastesnės kainos funkcijos išraiška:**</code><br>
Ši išraiška yra lygiavertė bet kompaktiškė.
    
\begin{multline*}
\text{Cost} (h_{\theta}(x), y) = -y\log(h_{\theta}(x)) - (1 -y)\log(1 - h_{\theta}(x))
\end{multline*}
</div>


Žinome, kad galimi tik du atvejai, y = 0 arba 1. <br>

<code>**Tarkime y = 1, tada gauname:**</code>
\begin{multline*}
\text{Cost} (h_{\theta}(x), y) = -log(h_{\theta}(x)) 
\end{multline*}

<code>**Tarkime y = 0, tada gauname:**</code>

\begin{multline*}
\text{Cost} (h_{\theta}(x), y) = -\log(1- h_{\theta}(x))
\end{multline*}

## Kryžminė entropija

<code>**Todėl logistinės regresijos kainos funkciją galime parašyti taip:**</code>

\begin{multline*}
J(\theta) = \dfrac{1}{m}\sum_{i=1}^{m}Cost(h_{\theta}(x^{(i)} ),y^{(i)}) = -\dfrac{1}{m}[ \sum_{i=1}^{m} y^{(i)}\log(h_{\theta}(x^{(i)})) + (1 -y^{(i)})\log(1 - h_{\theta}(x^{(i)}))]
\end{multline*}


<div class="alert alert-block alert-danger">

<code>**Kainos funkcijos vektorizuota forma:**</code>

\begin{multline*}
h =g(X\theta)
\end{multline*}
\begin{multline*}
J(\theta) = \dfrac{1}{m} \cdot(-y^{\top}log(h) - (1 - y)^{\top} log(1-h))
\end{multline*}
</div>


<code>**Maksimalios tikimybės įvertinimas (Maximum likelihood estimation):**</code>

Šią kainos funkciją galima gauti iš statistikos, naudojant [didžiausios tikimybės įvertinimo] principą (https://en.wikipedia.org/wiki/Maximum_likelihood_estimation)
Ši idėja statistikoje, kaip efektyviai rasti parametrus 𝜃 skirtingiems modeliams, taip pat turi gerą savybę, kad ji yra <code>**išgaubta (convex)**</code>. Ir iš esmės visi jį naudoja optomizuodami logistinės regresijos modelius. Ir nesijaudinkite dėl to, jei nežinote, kas yra maksimalios tikimybės įvertinimas. Tai tik gilesnis šios konkrečios kainos funkcijos pagrindimas .<br>

### MLE (maksimalios tikimybės įvertinimas)
<code>**MLE (maksimalaus tikimybės įvertinimas)**</code> nustato vidurkį ir dispersiją kaip parametrus nustatant konkrečias tam tikro modelio parametrų reikšmes. Šis parametrų rinkinys gali būti naudojamas numatant duomenis, reikalingus normaliam paskirstymui. MLE daro prielaidą, kad yra <kodas>**[jungtinė tikimybės masės funkcija (joint probability mass function)](https://www.youtube.com/watch?v=LJYGete9kQY)** </code>, o <code>**OLS (paprastieji mažiausi kvadratai)**</code> nereikalauja jokių stochastinių prielaidų, kad būtų sumažintas atstumas.

Dabar turime rasti parametrus 𝜃, kurie sumažintų J(𝜃). Taigi, jei bandytume tai sumažinti, gautume tam tikrą 𝜃 parametrų rinkinį.<br>

<code>**Optimizuoti 𝜃:**</code>

\begin{multline*}
\min_{\theta} J(\theta)
\end{multline*}

Galiausiai, jei bus pateiktas naujas duomenų pavyzdys su tam tikra savybių rinkiniu 𝑥, tada galime paimti 𝜃, kurias optimizavome su algoritmo treniruotėms skirtu duomenu rinkiniu, ir <code>**išvesti savo prognozes (hipotezė)**</code>.

<code>**išvesti prognozes su 𝑥:**</code>

\begin{multline*}
h_{\theta}(x) = \dfrac{1}{1+e^{-\theta^{\top}x}}
\end{multline*}

<code>**Išvesti yra tikimybė:**</code><br>
Mūsų hipotezė yra tikimybės, kad <code>**𝑦 yra 1 pagal 𝑥 kurie yra parametrizuotu su 𝜃**</code>

\begin{multline*}
P(y= 1 \vert x ; \theta) 
\end{multline*}

Belieka išsiaiškinti, kaip iš tikrųjų sumažinti J(𝜃) kaip 𝜃 funkciją, kad galėtume optimizuoti modelio 𝜃 parametrus.<br>

\begin{multline*}
\min_{\theta} J(\theta)
\end{multline*}

## Gradientinis nusileidimas 

<code>**Kainos funkcija:**</code>
\begin{multline*}
J(\theta) = -\dfrac{1}{m}[ \sum_{i=1}^{m} y^{(i)}\log(h_{\theta}(x^{(i)})) + (1 -y^{(i)})\log(1 - h_{\theta}(x^{(i)}))]
\end{multline*}

<code>**Mes norime:**</code>
\begin{multline*}
\min_{\theta} J(\theta)
\end{multline*}

<code>**Įprastas gradiento nusileidimo šablonas:**</code><br>
`repeat until convergence (simultaneously update all 𝜃j){
`\begin{multline*}
\theta_{j}:= \theta_{j} - \alpha \dfrac{\partial}{\partial\theta_{j}} J(\theta)
\end{multline*}
`}`

Čia yra mūsų įprastas gradiento nusileidimo šablonas, kuriame turime pakartotinai atnaujinti kiekvieną parametrą

<code>**Išvestiniai skaičiavimai pateikia šią lygtį:**</code>

\begin{multline*}
\dfrac{\partial}{\partial\theta_{j}} J(\theta) =  \dfrac{1}{m}\sum_{i=1}^{m}(h_{\theta} (x^{(i)})- y^{(i)}) \cdot x^{(i)}_{j}
\end{multline*}

Jei paimtume šį dalinį išvestinį terminą ir prijungtume jį atgal<br>

<code>**Gradientinis nusileidimas :**</code><br>

`repeat until convergence (simultaneously update all 𝜃j){
`\begin{multline*}
\theta_{j}: = \theta_{j} - \dfrac{\alpha}{m}\sum_{i=1}^{m}(h_{\theta} (x^{(i)})- y^{(i)}) \cdot x^{(i)}_{j}
\end{multline*}
`}`

Ši lygtis buvo būtent tokia, kokią turėjome tiesinei regresijai. Taigi natūralus klausimas ar  linijinės ir logistinės regresijos algoritmai skirtingi ar ne?<br>
Logistinės regresijos atveju kai kas pasikeitė tai <code>**hipotezės ℎ𝜃(𝑥) apibrėžimas.**</code>. Tuo tarpu tiesinei regresijai turėjome <code>**ℎ𝜃(𝑥) = 𝜃⊤𝑥**</code>, dabar ℎ𝜃(𝑥) apibrėžimas yra toks:

\begin{multline*}
h_{\theta}(x) = \dfrac{1}{1+e^{-\theta^{\top}x}}
\end{multline*}

So even though the update rule looks on surface identical because the definition of the hypothesis has changed this is actually not the same thing as gradient descent for linear regression.
Taigi, nors atnaujinimo taisyklė iš pažiūros atrodo identiška tai iš tikrųjų nėra tas pats, kas tiesinės regresijos gradiento nusileidimas nes pasikeitė hipotezės apibrėžimas.
_____________________________________________

Anksčiau trumpai kalbėjome apie tai, kaip stebėti gradiento nusileidimą, kad įsitikintumėte, jog jis konverguoja. Paprastai tas pats metodas taikomas logistinei regresijai, kad būtų galima stebėti GD ir įsitikinti, kad jis teisingai konverguoja.
Diegdami logistikos regresiją su gradiento nusileidimu, turime visas šias skirtingų parametrų reikšmes nuo 𝜃0, 𝜃1, 𝜃2 iki 𝜃n, kurias turime atnaujinti naudodami šią GD išraišką.

\begin{multline*}
\Theta = 
\begin{bmatrix}
\theta_{0} \\
\theta_{1} \\
\theta_{2} \\
. \\
. \\
. \\
\theta_{n}
\end{bmatrix}
\end{multline*}

## Logistinės regresijos taikymas

Galime turėti <code>**for ciklą, for (i=0 → n)**</code> bet idealiu atveju galėtume naudoti ir  <code>**vektorizuota formą**</code> kuri gali atnaujinti visus šiuos n+1 parametrus
 
\begin{multline*}
\theta: = \theta - \alpha\dfrac{1}{m}\sum_{i=1}^{m}[(h_{\theta} (x^{(i)})- y^{(i)}) \cdot x^{(i)}]
\end{multline*}


<div class="alert alert-block alert-danger">
    
<code>**Vektrozituoa notacija:**</code>
    
\begin{multline*}
\theta: = \theta - \dfrac{\alpha}{m}X^{\top}(g(X\theta) - \overrightarrow{y})
\end{multline*}
</div>

```python
# logistic function signature
def sigmoid(z):
    return 1/(1 + np.exp(-z))

# model function signature
def hypothesis(X,θ):
    z = X @ θ
    return sigmoid(z)

# vectorized cost function signature
def cost(h, y):
    m = len(y)
    return (1 / m) * ( np.sum( -y.T.dot(np.log(h)) - (1 - y).T.dot(np.log(1 - h)) ) )

# vectorized gradient descent function signature
def gradient(X, y, θ, α = 0.005, epochs=1000):
    m, n = X.shape
    history = np.zeros(epochs)
    for i in range(epochs):
        h          = hypothesis(X, θ)
        error      = h - y
        derivative = X.T @ error
        θ -= α/m * derivative
        history[i] = cost(h, y)
        
    return θ, history

# utility function for data plotting
def plot_cost(data, label='', xlabel='', ylabel=''):
    plt.figure(figsize=(12,8))
    plt.plot(range(len(data)), data,'r')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)

def predict(X, θ, p_thr = 0.5):
    proba       = hypothesis(X, θ)
    predictions = np.where(proba < 0.5, 0, 1)
    return predictions 
    # return np.array([1 if i > p_thr else 0 for i in y_pred]).reshape(-1,1)
    # return np.fromiter([1 if i > p_thr else 0 for i in y_pred], int).reshape(-1,1)
```
